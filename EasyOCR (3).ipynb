```python
import easyocr
import cv2
from google.colab.patches import cv2_imshow
reader = easyocr.Reader(['ar','en']) # this needs to run only once to load the model into memory
```

```python
img = cv2.imread("/content/arabic2.png")
result = reader.readtext(img)
print(result)
```

```python
for (bbox, text, acc) in result:
  (tl, tr, br, bl) = bbox
  tl = (int(tl[0]+10), int(tl[1]+10))
  tr = (int(tr[0]), int(tr[1]))
  br = (int(br[0]-10), int(br[1]-10))
  bl = (int(bl[0]), int(bl[1]))
  cv2.rectangle(img, tl, br, (0,0,0),3)

  # print(tl)
# tl = tl+(1,1)
# print(tl)

cv2_imshow(img)
```
